```python
import requests
import urllib3
from urllib import parse
import pandas as pd
import re

urllib3.disable_warnings()
url = "https://27.101.213.4/downloadCsvAge.do?searchYearMonth=month&xlsStats=1"
headers = {
    "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
    "accept-language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7,ja-JP;q=0.6,ja;q=0.5,la;q=0.4,zh-TW;q=0.3,zh-CN;q=0.2,zh;q=0.1",
    "cache-control": "max-age=0",
    "content-type": "application/x-www-form-urlencoded",
    "sec-fetch-dest": "document",
    "sec-fetch-mode": "navigate",
    "sec-fetch-site": "same-origin",
    "sec-fetch-user": "?1",
    "upgrade-insecure-requests": "1",
}

data = {
    "sltOrgType": "1",
    "sltOrgLvl1": "A", # A | 비우기 (계)
    "sltOrgLvl2": "A", # A | 비우기 (남녀구분)
    "gender": "", # gender | 비우기
    'sum': "sum",
    "sltUndefType": "",
    "searchYearStart": "2020",
    "searchMonthStart": "12",
    "searchYearEnd": "2020",
    "searchMonthEnd": "12",
    "sltOrderType": "1",
    "sltOrderValue": "ASC",
    "sltArgTypes": "5",# 나이간격 1 / 5 / 10
    "sltArgTypeA": "0",
    "sltArgTypeB": "100",
    "category": "year",# month / year
    "state": "1",
    "stateMobile": "1"
}

res = requests.post(url=url, headers=headers, data=data, verify=False, stream=True)
res.raise_for_status()
res.encoding = 'euc-kr'

headerFileName = res.headers["Content-Disposition"];
fileName = re.findall("filename=(.+)", parse.unquote(headerFileName))[0]
fileName = fileName.replace("\"","").replace("\"","").replace(";","")
with open("data/"+fileName, 'wb') as f:
    f.write(res.content)
df = pd.read_csv("data/"+fileName, encoding='euc-kr', thousands = ',')
```

```python
import math, random

exportNum = 4

def randPlace(n):
    total = 100
    result = []
    try:
        for i in range(n-1):
            expTotal = math.floor(total * 100)
            if expTotal <= 0:
                result.append(expTotal / 100)
                return result
            val = random.randint(1, expTotal) / 100
            result.append(val)
            total -= val
        result.append(math.floor(total * 100) / 100)
        return result
    except ValueError:
        print(math.floor(total * 100))

def export():
    List = randPlace(exportNum)
    Sum = 0
    for item in List:
        Sum += item
    if Sum == 100.0:
        return True, List
    else:
        return False, List

countLen = 0
count = 0
for time in range(100):
    checkSum = True
    while checkSum:
        store = export()
        if len(store[1]) != exportNum:
            countLen += 1
        if store[0] == True and len(store[1]) == exportNum:
            checkSum = False
        count += 1
print(count - countLen, countLen)
```
